# Next word prediction using LSTM

```python
import pandas as pd
import numpy as np
import pylab as plt

# NLP imports
import re
import nltk, gensim
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from tensorflow.keras.preprocessing.text import Tokenizer

data_dir = './data/'
```

```python
def prepare_word_data():
    file = open(data_dir + 'Metamorphosis_full_book.txt', 'r', encoding = "utf8")
    doc = ''
    for line in file.readlines():
        doc = doc + line

    REPLACE_BY_SPACE_RE = re.compile('[/(){}\[\]\|@,;\n]')
    BAD_SYMBOLS_RE = re.compile('[^0-9a-z #+_]')
    STOPWORDS = set(stopwords.words('english'))

    doc = doc.lower()
    doc = REPLACE_BY_SPACE_RE.sub(' ',doc)
    doc = BAD_SYMBOLS_RE.sub('', doc)
    doc = ' '.join([word for word in doc.split() if word not in STOPWORDS])
    return doc
```

```python
raw_doc = prepare_word_data()
raw_doc[1500:1600]
```

```python
def tokenize_doc(doc):
    doc = doc.split()
    stemmer = PorterStemmer()
    doc = [stemmer.stem(d) for d in doc]
    return doc

raw_doc = prepare_word_data()
doc_words = tokenize_doc(raw_doc)
doc_words[:5]
```

```python
def words_to_sequence(doc, n_steps=32):
    X, y = list(), list()
    for i in range(len(doc)-n_steps-1):
        X.append(doc[i:i+n_steps])
        y.append(doc[i+n_steps+1])
    return (X, y)

(X, y) = words_to_sequence(doc_words)
```
